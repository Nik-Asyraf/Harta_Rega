```python
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is your DataFrame containing the data
# Select the numerical columns you want to standardize
numerical_cols = ['Location', 'Bathrooms', 'Car Parks', 'Property Type', 'Size', 'Furnishing', 'Bedrooms', 'Store', 'Size_Encoded' , 'Price']

# Create a StandardScaler instance
scaler = StandardScaler()

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 5000)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
accuracy_percentage = accuracy * 100
print("Accuracy: {:.2f}%".format(accuracy_percentage))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
